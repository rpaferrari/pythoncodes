```python
from openpyxl import load_workbook as lwb
import os

nomeCaminhoArquivo = r'C:\Users\rpafe\Desktop\Secao 5 - Robo 4 - Preenchendo formulários WEB com Python e Excel\Input\DadosFormulario.xlsx'
planilha_aberta = lwb(nomeCaminhoArquivo) #ou lwb(filename=nomeCaminhoArquivo)

# Seleciona a sheet que tem as informações a serem passadas para o formulario on-line
sheet_selecionada = planilha_aberta['Dados']

for linha in range(2,len(sheet_selecionada['A']) + 1):
    nome = sheet_selecionada['A%s' % linha].value
    email = sheet_selecionada['B%s' % linha].value
    telefone = sheet_selecionada['C%s' % linha].value
    sexo = sheet_selecionada['D%s' % linha].value
    sobre = sheet_selecionada['E%s' % linha].value
    
    from selenium import webdriver as wbd
    from selenium.webdriver.common.keys import Keys
    from selenium.webdriver.common.by import By
    import pyautogui as pag

    navegadorFormulario = wbd.Chrome()
    navegadorFormulario.get(r'https://pt.surveymonkey.com/r/Y9Y6FFR')

    # Aguardar 1 segundo
    pag.sleep(1)

    # Preenche Nome
    navegadorFormulario.find_elements(By.XPATH, '//*[@id="683928983"]')[0].send_keys(nome)

    # Aguardar 1 segundo
    pag.sleep(1)

    # Preenche Email
    navegadorFormulario.find_elements(By.XPATH, '//*[@id="683932318"]')[0].send_keys(email)

    # Aguardar 1 segundo
    pag.sleep(1)

    # Preenche Telefone
    navegadorFormulario.find_elements(By.XPATH, '//*[@id="683930688"]')[0].send_keys(telefone)

    # Aguardar 1 segundo
    pag.sleep(1)

    # Preenche Escreva sobre você
    navegadorFormulario.find_elements(By.XPATH, '//*[@id="683932969"]')[0].send_keys(sobre)

    # Aguardar 1 segundo
    pag.sleep(1)

    # Preenche Sexo
    if sexo == 'Feminino':
        # Feminino
        navegadorFormulario.find_elements(By.ID, '683931881_4497366119_label')[0].click()
    else:
        # Masculino
        navegadorFormulario.find_elements(By.ID, '683931881_4497366118_label')[0].click()
    
    # Aguardar 1 segundo
    pag.sleep(1)

    # Aperta Botão Enviar
    navegadorFormulario.find_elements(By.XPATH, '//*[@id="patas"]/main/article/section/form/div[2]/button')[0].click()
```
