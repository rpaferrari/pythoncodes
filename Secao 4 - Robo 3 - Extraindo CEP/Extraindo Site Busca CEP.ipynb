```python
from openpyxl import load_workbook as lwb
import os

# Pega o caminho + nome do arquivo
nome_arquivo_cep = r'C:\Users\rpafe\Desktop\Secao 4 - Robo 3 - Extraindo CEP\Pesquisa Endereço\PesquisaEndereco_2.xlsx'
planilhaDadosEndereco = lwb(nome_arquivo_cep)

# seleciona a sheet da planilha de CEP
sheet_selecionada = planilhaDadosEndereco["CEP"]

from selenium import webdriver as wbd
from selenium.webdriver.common.keys import Keys
import pyautogui as pag

# Importando o elemento By
from selenium.webdriver.common.by import By

# Aguarda 1 segundo
pag.sleep(1)

# Abre o navegador do Google Chrome e abre o site do busca CEP
navegador = wbd.Chrome()
navegador.get("https://buscacepinter.correios.com.br/app/endereco/index.php")

# Aguarda 1 segundo
pag.sleep(1)

# Imprime o CEP no Campo do CEP
navegador.find_elements(By.NAME, r'endereco')[0].send_keys('13203532')
#navegador.find_element(By.NAME, r'endereco').send_keys("")

# Aguarda 1 segundo
pag.sleep(1)

# Clicando no botão btn_pesquisa lembrando que o nome pode mudar
navegador.find_element(By.NAME, "btn_pesquisar").click()

# Aguarda 1 segundo
pag.sleep(2)

for linha in range(2, len(sheet_selecionada['A']) + 1):

    # Aguarda 1 segundo
    pag.sleep(1)
    
    # Clicar nova busca pelo XPATH
    # navegador.find_element(By.XPATH, '//*[@id="btn_nbusca"]').click
    navegador.find_element(By.ID, 'btn_nbusca').click()
    
    # Aguarda 1 segundo
    pag.sleep(1)
    
    # Transforma a linha em String
    # linhaPesquisa = str(linha)
    cepPesquisa = sheet_selecionada['A%s' % linha].value
    
    # Aguarda 1 segundo
    pag.sleep(1)
    
    # Imprime o CEP no Campo do CEP
    navegador.find_elements(By.NAME, r'endereco')[0].send_keys(cepPesquisa)
    #navegador.find_element(By.NAME, r'endereco').send_keys("")

    # Aguarda 1 segundo
    pag.sleep(1)

    # Clicando no botão btn_pesquisa lembrando que o nome pode mudar
    navegador.find_element(By.NAME, 'btn_pesquisar').click()

    # Aguarda 1 segundo
    pag.sleep(2)
    
    # Insere a informação na variavel e imprime
    rua = navegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[1]')[0].text
    print(rua)

    # Insere a informação na variavel e imprime
    bairro = navegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[2]')[0].text
    print(bairro)

    # Insere a informação na variavel e imprime
    cidade = navegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[3]')[0].text
    print(cidade)

    # Insere a informação na variavel e imprime
    cep = navegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[4]')[0].text
    print(cep)
    
    # Aguarda 1 segundo
    pag.sleep(2)
    
    # Seleciona a Sheet de Dados
    sheet_Dados_Para_Imprimir_Endereco = planilhaDadosEndereco['Dados']
    linhaCorrentePlanilhaCEP = len(sheet_Dados_Para_Imprimir_Endereco['A']) + 1
    
    # Pegamos a última linha preenchida na coluna A e acrescentamos +1
    colunaA = "A" + str(linhaCorrentePlanilhaCEP) #criando variável para juntar A + a ultima linha exemplo: A2
    colunaB = "B" + str(linhaCorrentePlanilhaCEP) #criando variável para juntar B + a ultima linha exemplo: B2
    colunaC = "C" + str(linhaCorrentePlanilhaCEP) #criando variável para juntar C + a ultima linha exemplo: C2
    colunaD = "D" + str(linhaCorrentePlanilhaCEP) #criando variável para juntar D + a ultima linha exemplo: D2

    # imprimindo as informações do site na planilha
    sheet_Dados_Para_Imprimir_Endereco[colunaA] = rua
    sheet_Dados_Para_Imprimir_Endereco[colunaB] = bairro
    sheet_Dados_Para_Imprimir_Endereco[colunaC] = cidade
    sheet_Dados_Para_Imprimir_Endereco[colunaD] = cep

# Fecha o navegador
navegador.close()
    
# Salvando o arquivo do excel com as novas informações
planilhaDadosEndereco.save(filename=nome_arquivo_cep)

# Abrindo e apresentando na tela com os dados
os.startfile(nome_arquivo_cep)
```
