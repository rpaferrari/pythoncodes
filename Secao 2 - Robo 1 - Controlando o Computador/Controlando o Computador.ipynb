```python
########## Posição abre arquivos ##########
import pyautogui as pag

# Clicar com o mouse no frame indicado
pag.click(x=10, y=100)

# Hotkey, executa mais de uma tecla para atalhos
# Neste caso é a mesma coisa que precionar Windows + R
pag.hotkey('win', 'r')

# Tempo de espera
pag.sleep(1)

# Digitamos a palavra Excel
pag.typewrite('notepad')

# Apertamos a tecla enter
pag.press('Enter')

# Tempo de espera
pag.sleep(1)

# Digitamos a palavra
pag.typewrite('Abrimos o Notepad com um robô ou Script')

# getActiveWindow() permite pegar a janela que está ativa
fecharJanelaNotepad = pag.getActiveWindow()

# Tempo de espera
pag.sleep(1)

# Aciona o fechamento da janela
fecharJanelaNotepad.close()

# Tempo de espera
pag.sleep(1)

# Apertamos a tecla enter
pag.press('Tab')

# Tempo de espera
pag.sleep(1)

# Apertamos a tecla enter
pag.press('Enter')
```

```python
########## Abre o aplicativo que for escolhido ##########
import pyautogui as pag
import pyautogui as escolha_opcao

opcao = pag.confirm('Clique no botão desejado', buttons = ['Excel','Word','Notepad'])

if opcao == "Excel":
    escolha_opcao.hotkey('win','r')
    escolha_opcao.sleep(1)
    escolha_opcao.typewrite('Excel')
    escolha_opcao.sleep(1)
    escolha_opcao.hotkey('Enter')
    escolha_opcao.sleep(3)
    escolha_opcao.hotkey('Enter')
    escolha_opcao.typewrite("Escolheu Excel")

elif opcao == "Word":
    escolha_opcao.hotkey('win','r')
    escolha_opcao.sleep(1)
    escolha_opcao.typewrite('WinWord')
    escolha_opcao.sleep(1)
    escolha_opcao.hotkey('Enter')
    escolha_opcao.sleep(3)
    escolha_opcao.hotkey('Enter')

elif opcao == "Notepad":
    escolha_opcao.hotkey('win','r')
    escolha_opcao.sleep(1)
    escolha_opcao.typewrite('Notepad')
    escolha_opcao.sleep(1)
    escolha_opcao.hotkey('Enter')
    escolha_opcao.sleep(1)
    escolha_opcao.typewrite("Escolheu Notepad")
```
